```python
def get_info(data, feature):
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    total_outlier_num = ((data[feature] < (Q1 - 1.5 * IQR)) | (data[feature] > (Q3 + 1.5 * IQR))).sum()
    d = {'Mean' : [round(data[feature].mean(),2)] , 'Median' : [round(data[feature].median(),2)], 'Outlier' : [total_outlier_num]}
    df_info = pd.DataFrame(data = d, index = [f'{feature}'])
    return df_info
    
get_info(df, "charges")
```

## Pipeline

### import des librairies

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler, RobustScaler
from sklearn import set_config
set_config(display="diagram")
from sklearn.compose import ColumnTransformer, make_column_selector, make_column_transformer
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import validation_curve
import pickle
```

### Configuration de la Pipeline

```python
num_selector = make_column_selector(dtype_include=["int64", "float64"])
cat_selector = make_column_selector(dtype_exclude=["int64", "float64"])

cat_processor = OneHotEncoder(handle_unknown='ignore')
num_processor = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("standard", StandardScaler()) 
])

## Quand test avec StandardScaler / RobustScaler score plus bas
```

```python
preprocessor = make_column_transformer(
    (num_processor, num_selector), (cat_processor, cat_selector)
)
preprocessor
```

### Test sur Dataset 1

```python
model = make_pipeline(preprocessor, Lasso(alpha=19.9))
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
